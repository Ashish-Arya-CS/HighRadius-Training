### Ques 1) Find those groups which have more “True” values than “False” values in the below dataframe.

```python
# Creation of dataframe
import pandas as pd
import numpy as np
df1 = pd.DataFrame({'A': ['group1', 'group1', 'group2', 'group1','group2',
'group1', 'group2', 'group2','group2','group1'],
'B': ['true', 'true', 'true', 'false', 'false', 'false', 'false', 'true','false','true']})
df1
```

```python
# Creating function to implement our question without using loops
total_length = len(df1)
def compare(grp_name):
    temp = df1.loc[df1['A'] == grp_name]
    total_true = temp.loc[temp['B'] == 'true']
    total_false = temp.loc[temp['B'] == 'false']
    if len(total_true) > len(total_false):
        return True
    else:
        
        return False
cmpare = compare('group1')
if cmpare == True:
    print("group 1 has more true values than false values")
cmpare = compare('group2')
if cmpare == True:
    print("group 2 has more true values than false values")
```

## Ques 2)  a.Get the items not common to both series A and series B (Without using loops)

```python
import pandas as pd 
import numpy as np 

ser1 = pd.Series([1, 2, 3, 4, 5])
ser2 = pd.Series([4, 5, 6, 7, 8])

# union of the series 
union = pd.Series(np.union1d(ser1, ser2)) 

# intersection of the series 
intersect = pd.Series(np.intersect1d(ser1, ser2)) 

# uncommon elements in both the series 
notcommon = list(union[~union.isin(intersect)] )

print(notcommon) 
```

## Ques 2) b.Get the items common to both series A and series B(Without using loops).

```python
intersect = list(np.intersect1d(ser1, ser2)) 
print(intersect)
```

## Ques 3) a.Generate a random series of length 10 and find the positions of numbers that are multiples of 3 from a series?

```python
import pandas as pd
import numpy as np
num_series = pd.Series(np.random.randint(1, 100, 10))
print("Original Series:")
print(num_series)
num_series = np.array(num_series)
result = np.argwhere(num_series % 3==0)
print("Positions of numbers that are multiples of 3:")
print(result)
```

## Ques 3) b.Compute the cumulative difference between the consecutive number for the same series(without using loops).

## Ques 1) Find the Average price of project from each state.

```python
# Reading the csv file
import pandas as pd
import numpy as np
df = pd.read_csv("Grouby_Assignment_Data.csv")
df.head(10)
```

```python
# Dropping the extra 'Unnamed: 0.1' column
df.drop(['Unnamed: 0.1'], axis=1, inplace=True)
df.head(10)
```

```python
# Finding The Average price of project from each state.
h = pd.DataFrame(df['price'].groupby(df['school_state']).mean())
h.rename(columns = {'price':'average_price'}, inplace = True) 
print("The Average price of project from each state :")
print(h)
```

## Ques 2) a. Find the total number of projects previously posted by all the teachers belonging to each teacher prefix.
### For Example all the teachers having prefix as dr have posted a total of 13 projects combined previously.

```python
prev = pd.DataFrame(df['teacher_number_of_previously_posted_projects'].groupby(df['teacher_prefix']).count())
print("The total number of projects previously posted by all the teachers belonging to each teacher prefix: \n")
print(prev)
```

## Ques 2) b. Find the prefix of the teacher who has posted the maximum of projects previously.

```python
highest_project = pd.DataFrame(prev[prev.teacher_number_of_previously_posted_projects == prev.teacher_number_of_previously_posted_projects.max()])
print("The prefix of the teacher who has posted the maximum of projects previously :")
highest_project
```

## Ques 5)Find the desired output as shown in the image below. Note : Create the data frame (Train Data) as shown in the image below (Output can be in fraction form or decimal form )

```python
# Creating the dataframe as given in image in the question.
df3 = pd.DataFrame({"State":['A','B','C','A','A','B','A','A','C','C'],'class': [0,1,1,0,1,1,0,1,1,0]})
grp_by = df3.groupby('class')
g0 = grp_by.get_group(0)
g1 = grp_by.get_group(1)
count0 = g0["State"].value_counts()
count1 = g1["State"].value_counts()

final0 = count0.divide(count1.add(count0,fill_value=0),fill_value = 0)
final1 = count1.divide(count1.add(count0,fill_value=0),fill_value = 0)

dff0 = pd.DataFrame({"State":list(final0.index),"State_0":list(final0)})
dff1 = pd.DataFrame({"State":list(final1.index),"State_1":list(final1)})

# Printing the desired output.
print(df3.merge(dff0,left_on = "State", right_on = "State").merge(dff1,left_on = "State", right_on = "State"))
```
